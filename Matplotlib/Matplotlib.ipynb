# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# FUNCTIONAL
plt.plot(x, y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

```python
# Object-Oriented
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # [left, bottom, width, height]
axes.plot(x, y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x, y)
axes2.plot(y, x)
axes2.set_title('Smaller Plot')
axes1.set_title('Larger Plot')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

# axes.plot(x, y)

axes[0].plot(x, y)
axes[0].set_title('First Plot')

axes[1].plot(y, x)
axes[1].set_title('Second Plot')
```

# Figure Size and DPI

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.set_title('Title')
ax.set_ylabel('Y Label')
ax.set_xlabel('X Label')


ax.plot(x, x**2, label='X Squared')
ax.plot(x, x**3, label='X Cubed')

ax.legend()
```

# Saving a Figure

```python
fig.savefig('my_picture.png')
```

# Plot Appearance

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, color='green', linewidth=2, ls='--')

ax.set_xlim([0, 1])
ax.set_ylim([0, 2])
```
